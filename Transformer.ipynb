```python
import torch
num_tokens=mytokenizer.get_token_num()
dim_model=1024
num_heads=16
num_encoder_layers=12
num_decoder_layers=12
dropout_p=0
transmodel=MolTransformer(num_tokens,dim_model,num_heads,num_encoder_layers,
                          num_decoder_layers,dropout_p).to(DEVICE)
```

Train

```python
from torch import nn
import matplotlib.pyplot as plt

optimizer=optim.Adam(transmodel.parameters(),lr=0.000001,betas=(0.99,0.995))
criteron=nn.CrossEntropyLoss(ignore_index=mytokenizer.get_token_encoding('<pad>'))
```

```python
import numpy as np
def get_padding_mask(seq):
  pad_attn_mask=seq.data.eq(0)
  return pad_attn_mask

def get_target_subsequent_mask(seq):
  attn_shape=[seq.size(0),seq.size(1),seq.size(1)]
  mask=np.triu(np.ones(attn_shape),k=1)
  return torch.from_numpy(mask).byte()
```

```python
def get_character(vec):
  top5_value, top5_idx=vec.topk(3,dim=2)
  top5_tokens=mytokenizer.get_inv_mapping(top5_idx)
  return top5_value,top5_tokens

```

```python
from traitlets.traitlets import validate
Epoch=5
x = []
t_loss = []
step=0
torch.set_printoptions(profile="short")
for epoch in range(Epoch):
  for i, data in tqdm(enumerate(train.dataset),total=len(train.dataset)):
    src,dec_in,tgt=data
    optimizer.zero_grad()
    #transmodel.zero_grad()
    src=src.to(DEVICE)
    dec_in=dec_in.to(DEVICE)
    tgt=tgt.to(DEVICE)
    src_pad_mask=get_padding_mask(src)
    dec_pad_mask=get_padding_mask(dec_in)
    #print(mytokenizer.get_inv_mapping(src))
    #print(src_pad_mask)
    #tgt_mask=get_target_subsequent_mask(dec_in)
    out=transmodel(src,dec_in,src_pad_mask,dec_pad_mask)
    #print(out.shape)
    #print(out[0,1:10])
    #vis=get_character(out)
    #print(vis)
    loss=criteron(out.transpose(1,2),tgt)
    loss.backward()
    optimizer.step()
    
    
    x.append(step) 
    step+=1
    t_loss.append(loss.cpu().detach().numpy())
    if len(x)>2000:
        x.pop(0)
        t_loss.pop(0)

    if i%20==0:
      try:
        train_loss_lines.remove(train_loss_lines[0])
      except Exception:
        pass
      train_loss_lines = plt.plot(x, t_loss, 'r', lw=1)

      plt.title("loss")
      plt.xlabel("steps")
      plt.ylabel("loss")
      plt.legend(["train_loss"])
      plt.pause(0.1) #stay0.1s
      
    if i%100==0:
      top3_value,top3_atom=get_character(out)
      print(f'epoch: {epoch}')
      print(f'value: {top3_value}')
      print(f'atom: {top3_atom}')
      print(f'loss is {loss}')
    #if i%1000==0:
    #  random_ids = np.random.randint(len(validate), size=1)
    #  select_batch=validate[random_ids]
```
